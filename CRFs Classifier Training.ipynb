###clean data

```python
def normalize_text(content):
    text = re.sub(r"http\S+", "", content ) # Remove urls
    text = text.lower() # Lowercase 
    text = text.strip() # Remove leading/trailing whitespace
    text = re.sub('\s+', ' ', text) # Remove extra space and tabs
    text = re.sub('\n',' ',text) # remove newlines
    text = re.compile('<.*?>').sub('', text) # Remove HTML tags/markups:
    return text
```

```python
# Use a tokenizer from the NLTK library
import nltk
from nltk.tokenize import word_tokenize
def Stopword_removal(text):
  filtered_sentence = []

  # Stopword lists can be adjusted for your problem
  stopwords = ["a", "an", "the", "this", "that", "is", "it", "to", "and"]

  # Tokenize the sentence
  words = word_tokenize(text)
  for w in words:
      if w not in stopwords:
          filtered_sentence.append(w)
  text = " ".join(filtered_sentence)
  return text
```

```python
%%time
df['Resume_str_clean'] = df['Resume_str'].apply(normalize_text)
df['Resume_str_clean'] = df['Resume_str_clean'].apply(Stopword_removal)
```

### find POS_tag of each word 

```python
%%time
def pos_tags(document):
        sentences = nltk.sent_tokenize(document)
        print(len(sentences))
        sentences = [nltk.word_tokenize(sent) for sent in sentences]        
        sentences = [nltk.pos_tag(sent) for sent in sentences]
        
        return sentences
```

```python
%%time
training = pos_tags(training)
test = pos_tags(test)
```

### give lable to data 

```python
#without Stemming v4 after optimization2
def give_label(x):
  for i in range(len(x)):
    for j in range(len(x[i])):
        if len(x[i][j])<3 and x[i][j][0].lower() in df1['Skill_1'].values:                   
            if j<len(x[i])-1 and x[i][j+1][0].lower() in df1[df1['Skill_1']==x[i][j][0].lower()] ['Skill_2'].values:  
                if j<len(x[i])-2 and x[i][j+2][0].lower() in df1[df1['Skill_1']==x[i][j][0].lower()] ['Skill_3'].values:  
                  x[i][j]=x[i][j]+("B-skill",)
                  x[i][j+1]= x[i][j+1]+("I-skill",)
                  x[i][j+2]= x[i][j+2]+("I-skill",)
                elif (df1[(df1['Skill_1']==x[i][j][0].lower()) & (df1['Skill_2']==x[i][j+1][0].lower())]['Skill_3'].values == None).any() :
                  x[i][j]=x[i][j]+("B-skill",)
                  x[i][j+1]= x[i][j+1]+("I-skill",)
                elif (df1[df1['Skill_1']==x[i][j][0].lower()]['Skill_2'].values == None).any() :
                  x[i][j]=x[i][j]+("B-skill",)
                else :
                  x[i][j]=x[i][j]+("O",) 

            elif  (df1[df1['Skill_1']==x[i][j][0].lower()]['Skill_2'].values == None).any() :
               x[i][j]=x[i][j]+("B-skill",)
            else :
               x[i][j]=x[i][j]+("O",) 
        elif len(x[i][j])<3:
          x[i][j]=x[i][j]+("O",) 
                           
             
  return x
```

### Feature Engineering

```python
%%time
def word2features(sent, i):
    word = sent[i][0]
    postag = sent[i][1]
    features = {
        'bias': 1.0,
        'word.lower()': word.lower(),
        'word[-3:]': word[-3:],
        'word[-2:]': word[-2:],
        'word.isupper()': word.isupper(),
        'word.istitle()': word.istitle(),
        'word.isdigit()': word.isdigit(),
        'postag': postag,
        'postag[:2]': postag[:2],
    }
    if i > 0:
        word1 = sent[i-1][0]
        postag1 = sent[i-1][1]
        features.update({
            '-1:word.lower()': word1.lower(),
            '-1:word.istitle()': word1.istitle(),
            '-1:word.isupper()': word1.isupper(),
            '-1:postag': postag1,
            '-1:postag[:2]': postag1[:2],
        })
    else:
        features['BOS'] = True
    if i < len(sent)-1:
        word1 = sent[i+1][0]
        postag1 = sent[i+1][1]
        features.update({
            '+1:word.lower()': word1.lower(),
            '+1:word.istitle()': word1.istitle(),
            '+1:word.isupper()': word1.isupper(),
            '+1:postag': postag1,
            '+1:postag[:2]': postag1[:2],
        })
    else:
        features['EOS'] = True
    return features
def sent2features(sent):
    return [word2features(sent, i) for i in range(len(sent))]
X_train = [sent2features(s) for s in training]
X_test = [sent2features(s) for s in test]
```

```python
%%time
def sent2labels(sent):
    return [label for token, postag, label in sent]
y_train = [sent2labels(s) for s in training]
y_test = [sent2labels(s) for s in test]
```

### train  a CRF

```python
%%time
crf = sklearn_crfsuite.CRF(
    algorithm='lbfgs',
    c1=0.1,
    c2=0.1,
    max_iterations=100,
    all_possible_transitions=True
)
#crf.fit(X_train, y_train)
try:
    crf.fit(X_train, y_train)
except AttributeError:
    pass
```

```python
labels = list(crf.classes_)
print(labels)
```

### Predict 

```python
%%time
y_pred = crf.predict(X_test)
```

```python
#to get list of all skill with confidance score
def get_skills_ScoreV3(text):
    text = pos_tags(text)
    z=text
    text=[sent2features(s) for s in text] 
    l=[]
    for i in range(len(z)):
      for j in range(len(z[i])):
          score=crf.predict_marginals_single(text[i])[j]
          lable=max(score,key=score.get)
          lable_score=max(score.values())
        
          if lable=="B-skill":
            score2=crf.predict_marginals_single(text[i])[j+1]
            lable2=max(score2,key=score2.get)
            lable_score2=max(score2.values())
            if j<len(z[i])-2 and lable2=="I-skill":
              if j<len(z[i])-3:
                score3=crf.predict_marginals_single(text[i])[j+3]
                lable3=max(score3,key=score3.get)
                lable_score3=max(score3.values())
                if lable3=="I-skill":
                  skill=z[i][j][0]+' '+z[i][j+1][0]+' '+z[i][j+2][0]
                  final_score=lable_score*lable_score2*lable_score3
                  l.append({skill:final_score})
                else:
                  skill=z[i][j][0]+' '+z[i][j+1][0]
                  final_score=lable_score*lable_score2
                  l.append({skill:final_score}) 
            else:
                skill=z[i][j][0]
                final_score=lable_score
                l.append({skill:final_score})
    print(l) 
```
